```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy.stats as stats


data = pd.read_excel('assets/day-2-prob-4.xlsx')
```

```python
# Draw 2000 random samples (with replacement) of size 50 from the Height_cm column, compute their means, and plot the sampling distribution:

np.random.seed(42)
sample_means = []
for _ in range(2000):
    sample = data['Height_cm'].sample(n=50, replace=True)
    sample_means.append(sample.mean())

plt.figure(figsize=(8,5))
sns.histplot(sample_means, bins=20, color='salmon', kde=True, stat='count', edgecolor='black', alpha=0.4)
plt.title('Sampling Distribution of the Mean (CLT in Action)')
plt.xlabel('Sample Mean Height (Cm) (n=50)')
plt.ylabel('Count')
plt.show()
```
